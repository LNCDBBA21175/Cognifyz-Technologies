```python
#Task 2 :-
import pandas as pd
import matplotlib.pyplot as plt

file_path = "c"
df = pd.read_csv("C:\\sahil\\project_folder\\cognifyz task 1\\Data_set 2 - Copy.csv")

# Step 1: Extract gender column
gender_counts = df['gender'].value_counts()

# Step 2: Create bar chart
plt.figure(figsize=(6,4))
gender_counts.plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Step 2 : Create pie chart
plt.figure(figsize=(6,6))
gender_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
plt.title("Gender Distribution")
plt.ylabel("")
plt.show()
```

```python
#Task 3 :-
import pandas as pd

df = pd.read_csv("C:\\sahil\\project_folder\\cognifyz task 1\\Data_set 2 - Copy.csv")

# 1. Identify Numerical Columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
print("Numerical Columns:", numerical_cols)

# 2. Calculate Statistics: mean, median, std
basic_stats = df[numerical_cols].agg(['mean', 'median', 'std']).transpose()

# Display results
print("\nBasic Statistics for Numerical Columns:\n")
print(basic_stats)

basic_stats.to_csv("basic_statistics.csv")
```

```python
# Task 5 :-
import pandas as pd

df = pd.read_csv("C:\\sahil\\project_folder\\cognifyz task 1\\Data_set 2 - Copy.csv")

# Select all reason columns
reason_cols = ['Reason_Equity', 'Reason_Mutual', 'Reason_Bonds', 'Reason_FD']

# Combine into one column
all_reasons = df[reason_cols].apply(lambda x: ' '.join(x.dropna().astype(str)), axis=1)

# Flatten and count frequencies
reason_counts = all_reasons.str.split(expand=True).stack().value_counts()

print("Most common reasons for investment:")
print(reason_counts.head(10))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("C:\\sahil\\project_folder\\cognifyz task 1\\Data_set 2 - Copy.csv")

# Reason columns
reason_cols = ['Reason_Equity', 'Reason_Mutual', 'Reason_Bonds', 'Reason_FD']

# Plot top 3 reasons for each column
for col in reason_cols:
    plt.figure(figsize=(6,4))
    df[col].value_counts().head(5).plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title(f"Top Reasons for {col.replace('Reason_', '')} Investment")
    plt.xlabel("Reason")
    plt.ylabel("Count")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()
```
